```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plot
import seaborn as sns
```

```python
data = pd.read_csv("tips.csv")
data
```

```python
tips_stats = data['tip'].describe()
tips_stats
```

```python
# Addtional Statistics

mean_tip = data['tip'].mean()
median_tip = data['tip'].median()
std_tip = data['tip'].std()
var_tip = data['tip'].var()
skew_tip = data['tip'].skew()
kurtosis_tip = data['tip'].kurtosis()
cv_tip = (std_tip/mean_tip)*100
q1=np.percentile(data['tip'], 25)
q2 = np.percentile(data['tip'], 50)
q3 = np.percentile(data['tip'], 75)
print("\nAdditional statistics for the tip column:")
print("Mean :", mean_tip)
print("Median :", median_tip)
print("Standard Deviation :", std_tip)
print("Variance :", var_tip)
print("Skewness :", skew_tip)
print("Kurtosis :", kurtosis_tip)
print("Coefficient of variation :", cv_tip)
print("First Quartile (q1) :", q1)
print("Second Quartile (q2) :", q2)
print("Third Quartile (q3):", q3)

```
